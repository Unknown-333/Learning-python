# Using list comprehension

````python
fruits = ['Apple', 'Banana', 'Apple', 'Mango', 'Guava', 'Papaya', 'Kiwi', 'Orange']
````

````python
[item.lower() for item in fruits]
````

````python
# Use for loop and list comprehension both to find square of numbers between 1 to 10
num = []
for i in range(1,11):
    num.append(i**2)
print(num)
````

````python
sqr = [i ** 2 for i in range(1,11)]
print(sqr)
````

````python
# find even number between 1 to 25 using list comprehension
even = [j for j in range(2,26,2)]
print(even)
````
